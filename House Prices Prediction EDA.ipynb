# Data Preprocessing


The 'id','zipcode' columns will be droped because each property has its unique id, and zipcode.


```python
data.drop("id",axis=1,inplace=True)
data.drop("zipcode",axis=1,inplace=True)
```

```python
data.head()
```

"time" column is splitted into three related columns: "year", "month", "day", after that "time" column is deleted. 

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
# earliest date
data['date'].min()
```

```python
# latest date
data['date'].max()
```

```python
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data.head()
```

```python
data = data.drop("date",axis=1)
data.head()
```

# Data Exploration

This section will study the hypotheses that related to the price 

```python
# Hypothesis 1: increasing number of bedrooms would increase the price 
sns.barplot(data=data, x="bedrooms", y="price")
plt.title("Relationship between number of bedrooms & average price ")
```

The graph shows a relationship between the number of bedrooms and the average price, but we need another chart to understand the relationship more clearly.

```python
sns.lineplot(data=data, x="bedrooms", y="price",err_style=None)
plt.title("Relationship between number of bedrooms & price ")
```

Line graph doesn't show a clear relationship between the number of bedrooms and the average price. For that, a box plot chart has been chosen to take a clear vision on the relationship.

```python
sns.boxplot(data=data, x="bedrooms", y="price",sym="")
```

The graph shows that there is a positive relationship between bedrooms & price. (more bedrooms means higher price, which is confirming the hypothesis)
